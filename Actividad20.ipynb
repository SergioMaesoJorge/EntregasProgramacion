```python
import pandas as pd
import matplotlib.pyplot as plt
datos=pd.read_csv("products.csv")
df=datos[["categoryID","unitsInStock"]]
valor = df.groupby("categoryID")["unitsInStock"].mean()
valor.head(15).plot.barh()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
datos=pd.read_csv("products.csv")
#print(datos)
df=datos[["categoryID","unitPrice"]]
valor = df.groupby("categoryID")["unitPrice"].mean()
df.head(15).categoryID.plot.pie()
plt.show()
```
